```python
import xlrd
```

```python
wb = xlrd.open_workbook("./SupplementalData.xlsx")
sheet = wb.sheet_by_index(0)
```

```python
for i in range(sheet.nrows):
    temp = ""
    for y in range(sheet.ncols):
        if(y == sheet.ncols-1):
            temp += str(sheet.cell_value(i, y))+ "\n"
        else:
            temp += str(sheet.cell_value(i, y))+ ", "
    print(temp)
```
